```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
# Plotting a Vector
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5, angles='xy', scale_units='xy', scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
# Vector Addition
vect1=np.asarray([0,0,4,5])
vect2=np.asarray([0,0,4,-3])
vect3=vect1+vect2
print(vect3)
```

```python
plt.quiver(0,0,4,5, angles='xy', scale_units='xy', scale=1, color='b')
plt.quiver(0,0,4,-3, angles='xy', scale_units='xy', scale=1, color='g')
plt.quiver(0,0,8,2, angles='xy', scale_units='xy', scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
# Vector Subtraction
vect1=np.asarray([0,0,4,5])
vect2=np.asarray([0,0,-4,3])
vect3=vect1-vect2
print(vect3)
```

```python
plt.quiver(0,0,4,5, angles='xy', scale_units='xy', scale=1, color='b')
plt.quiver(0,0,-4,3, angles='xy', scale_units='xy', scale=1, color='g')
plt.quiver(0,0,8,2, angles='xy', scale_units='xy', scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
# Vector Multiplication
vect1=np.asarray([0,0,4,5])
vect2=2*vect1
print(vect2)
```

```python
plt.quiver(0,0,4,5, angles='xy', scale_units='xy', scale=1, color='b')
plt.quiver(0,0,8,10, angles='xy', scale_units='xy', scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-2,12)
plt.show()
```
