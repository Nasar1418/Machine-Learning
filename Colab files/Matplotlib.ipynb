```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
plt.plot(x,z)
plt.xlabel('angle')
plt.ylabel('cosine value')
plt.title('cosine wave')
plt.show()
```

```python
plt.plot(x,y,'r-')
plt.plot(x,z,'g--')
plt.show()
```

```python
#parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y,'rx')
plt.show()
```

```python
#Bargraph
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
subjects=["Telugu","English","Maths","Science","Social"]
marks=[90,35,70,45,55]
ax.bar(subjects,marks)
plt.xlabel("Languages")
plt.ylabel("Marks")
plt.title("Marks of Students")
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
subjects=["Telugu","English","Maths","Science","Social"]
marks=[90,35,70,45,55]
ax.pie(marks,labels=subjects,autopct="%1.1f%%")
plt.show()
```

```python
#scatter plot
x=np.linspace(0,10,40)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color="b")
ax.scatter(x,z,color="r")
plt.show()
```

```python
#3d Scatter plot
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
fig3=plt.figure()
ax=plt.axes(projection="3d")
ax.scatter(x,y,z,c=z,cmap="Reds")
plt.show()
```
