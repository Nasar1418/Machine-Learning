```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# total bill vs tips
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.set_theme()
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker", style="smoker", size="size")
```

```python
# Load iris dataset
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.scatterplot(data=iris, x="sepal_length", y="petal_length", hue="species")
```

```python
sns.scatterplot(data=iris, x="sepal_length", y="petal_width", hue="species", style="species")
```

```python
# Load Titanic dataset
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
sns.countplot(data=titanic, x="survived")
```

```python
sns.countplot(data=titanic, x="class")
```

```python
# Bar Chart
sns.barplot(data=titanic, x="sex", y="survived", hue="class")
```

```python
# House Price Dataset
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
print(housing)
```

```python
housing_df=pd.DataFrame(housing.data, columns=housing.feature_names)
housing_df['price']=housing.target
housing_df.head()
```

```python
sns.distplot(housing_df['price'])
```

```python
correlation=housing_df.corr()
print(correlation)
```

```python
plt.figure()
sns.heatmap(correlation, annot=True, fmt="1.1f")
plt.show()
```
